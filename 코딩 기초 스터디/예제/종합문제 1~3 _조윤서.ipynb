```python
def GuGu(n):
  a = []
  for i in range(1, 10):
    a.append(n * i)
  print(a)
GuGu(2)
```

```python
result = 0
num = int(input('수를 입력하세요. :'))
for i in range (1, num+1):
  if i % 3 == 0 or i % 5 == 0:
    result += i
print(result)
```

```python
def GetTotalPage(m, n):
  if m%n == 0:
    result = m/n
  else:
    result = m//n +1
  print('총 페이지 수 :',int(result))

GetTotalPage(30,10)
```

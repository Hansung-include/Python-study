# 예외처리 _오류 일부러 발생시키기

```python
class Bird:
  def fly(self):
    raise NotImplementedError
#raise 명령어를 사용해 오류를 강제로 발생

class Eagle(Bird):
  def fly(self):
    print('very fast')
#메서드 오버라이딩을 통해 자식클래스 수정

eagle = Eagle()
print(eagle.fly())
```

# _sorted (ver.웅이)

```python
phone_book = {'홍길동': '010-1234-5678', '강감찬': '010-1234-5679'} #딕셔너리 생성
print(phone_book.items())
sorted_phone_book1 = sorted(phone_book.items(), key = lambda x: x[0])
#key값을 기준으로 비교를 하여 정렬
print(sorted_phone_book1)
```

# 갑분 헷갈렸던 global전역/지역변수

```python
result = 0
def add(num):
  global result 
#전역변수에 영향을 끼치기 위해 global사용!
#함수안에서의 일은 안에서 묻고 가기때문에..
  result += num
  return result

print(add(3))
print(add(4))
```

```python
class FourCal:
  

  def __init__(self, first, second):
#__init__ : 생성자, class로 만든 instance를 찍어낼때 얘를 "실행"하면서 시작됨
#생성자를 "실행"하려면 first, second가 필요
    self.first = first
    self.second = second

#하위 애들은 정의만 해놓은 것, 실행X
  def setdata(self, first, second):
    self.first = first
    self.second = second
  def add(self):
    result = self.first + self.second
    return result
 
  def div(self):
    result = self.first / self.second
    return result


class SafeFourCal(FourCal):
  def div(self):
    if self.second == 0:
      print('limitation')
    else:
      return self.first / self.second
#매서드 오버라이딩 : 부모의 기능을 수정. 자식 이기는 부모 없다

a = FourCal(1,2)
b = SafeFourCal(4,0)
print(b.div())
#???왜 None이 나오지....?
c = 10
c.add(1,2)
```
